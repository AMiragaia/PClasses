```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import json

# Function to load data
def load_data(filename):
    df = pd.read_excel(filename, skiprows=7)  # jumps the first 7 rows
    return df

# Function to clean data and extract unemployment rate
def clean_data(df):
    # Rename the first cell of the first column to 'Dados'
    df.iat[0, 0] = 'Dados'

    # Select the row with 'Taxa de desemprego (%)'
    unemployment_data = df.loc[df['Dados'] == 'Taxa de desemprego (%)']

    # Transpose the table and reset the index
    unemployment_data = unemployment_data.transpose()
    unemployment_data.reset_index(inplace=True)

    # Rename columns
    unemployment_data.columns = ['Year', 'Unemployment Rate']

    # Remove the first row that now contains 'Dados' and 'Taxa de desemprego (%)'
    unemployment_data = unemployment_data.iloc[1:]

    return unemployment_data

# Function to write DataFrame to a JSON file
def write_to_json(df, filename):
    # Convert DataFrame to dictionary and then write that dictionary to a JSON file
    data_dict = df.to_dict('records')
    with open(filename, 'w') as f:
        json.dump(data_dict, f)

# Function to visualize results
def visualize_results(df):
    plt.plot(df['Year'], df['Unemployment Rate'])
    plt.title('Unemployment Rate Evolution in Portugal')
    plt.xlabel('Year')
    plt.ylabel('Unemployment Rate (%)')
    plt.show()

# Main function
def main():
    # Load data
    filename = 'dados.xls'
    data = load_data(filename)

    # Clean data
    cleaned_data = clean_data(data)

    # Save cleaned data to a JSON file
    write_to_json(cleaned_data, 'unemployment_data.json')

    # Visualize results
    visualize_results(cleaned_data)

# Run the main function
if __name__ == '__main__':
    main()
```

```python
# Import necessary libraries
import pandas as pd
import json
import matplotlib.pyplot as plt
import numpy as np

# Function to load data
def load_data(filename):
    df = pd.read_excel(filename, skiprows=7)  # jumps the first 7 rows
    return df

# Function to clean data
def clean_data(df):
        # Find the index of the row with the desired value in the first cell
    df = df.iloc[:, :-2]  # Remove the last two columns
    row_index = df[df.columns[0]] == 'Taxa de desemprego (%) '
    
    # Select the row based on the index found
    unemployment_data = df.loc[row_index]

    # Drop the first column
    unemployment_data = unemployment_data.drop(unemployment_data.columns[0], axis=1)

    # Transpose the data
    unemployment_data = unemployment_data.transpose().reset_index()
    print(unemployment_data)

    return unemployment_data




# Function to write DataFrame to a JSON file
def write_to_json(df, filename):
    df.to_json(filename)

# Main function
def main():
    # Load data
    filename = 'dados.xls'
    data = load_data(filename)

    # Clean data
    cleaned_data = clean_data(data)

    # Save cleaned data to a JSON file
    write_to_json(cleaned_data, 'unemployment_data.json')

    # Assume 'cleaned_data' is the DataFrame containing the data
    # Set the index as the x-axis
    x = cleaned_data['index'].values
    
    # Set the values as the y-axis
    y = cleaned_data.iloc[:, 1].values
    
    # Plot the data as a line graph
    plt.plot(x, y, marker='o')
    
    # Set the y-axis limits
    plt.ylim(0, max(y) + 1)  # Add a buffer of 1 to the maximum value

    # Set the x-axis tick locations and labels
    plt.xticks(x, x.astype(int))
    
    # Add labels and title
    plt.xlabel('Year')
    plt.ylabel('Unemployment Rate (%)')
    plt.title('Unemployment Rate Over the Years')
    
    # Display the graph
    plt.show()




# Run the main function
if __name__ == '__main__':
    main()
```
