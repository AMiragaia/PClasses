```python
# Import necessary libraries
import pandas as pd
import json
import matplotlib.pyplot as plt
import numpy as np

# Function to load data
def load_data(filename):
    df = pd.read_excel(filename, skiprows=7)  # jumps the first 7 rows
    return df

# Function to clean data
def clean_data(df):
        # Find the index of the row with the desired value in the first cell
    df = df.iloc[:, :-2]  # Remove the last two columns
    row_index = df[df.columns[0]] == 'Taxa de desemprego (%) '
    
    # Select the row based on the index found
    unemployment_data = df.loc[row_index]

    # Drop the first column
    unemployment_data = unemployment_data.drop(unemployment_data.columns[0], axis=1)

    # Transpose the data
    unemployment_data = unemployment_data.transpose().reset_index()
    print(unemployment_data)

    return unemployment_data

def clean_data1(df):
    # Select the desired rows and columns until the year 2019
    data = df.iloc[8:35, [0, 1]]
    print(data)


    return data





# Function to write DataFrame to a JSON file
def write_to_json(df, filename):
    df.to_json(filename)

# Main function
def main():
    # Load data
    filename = 'dados.xls'
    data = load_data(filename)
    filename1 = 'emprego.xlsx'
    data1 = load_data(filename1)

    # Clean data
    cleaned_data = clean_data(data)
    cleaned_data1 = clean_data1(data1)

    # Save cleaned data to a JSON file
    write_to_json(cleaned_data, 'unemployment_data.json')

    # Assume 'cleaned_data' is the DataFrame containing the data
    # Set the index as the x-axis
    x = cleaned_data['index'].values
    
    # Set the values as the y-axis
    y = cleaned_data.iloc[:, 1].values
    
    # Plot the data as a line graph
    plt.plot(x, y, marker='o')
    
    # Set the y-axis limits
    plt.ylim(0, max(y) + 1)  # Add a buffer of 1 to the maximum value

    # Set the x-axis tick locations and labels
    plt.xticks(x, x.astype(int))
    
    # Add labels and title
    plt.xlabel('Year')
    plt.ylabel('Unemployment Rate (%)')
    plt.title('Unemployment Rate Over the Pandemic Years')
    
    # Display the graph
    plt.show()

    # Assume 'cleaned_data' is the DataFrame containing the data
    x = cleaned_data1['Unnamed: 0'].values
    y = cleaned_data1.iloc[:, 1].values
    
    # Plot the data as a line graph
    plt.plot(x, y, marker='o')
    
    # Set the y-axis limits
    plt.ylim(0, max(y) + 1)  # Add a buffer of 1 to the maximum value
    
    # Set the x-axis tick locations and labels
    plt.xticks(x, x.astype(int))
    plt.xticks(rotation=45)
    
    # Add labels and title
    plt.xlabel('Year')
    plt.ylabel('Unemployment Rate (%)')
    plt.title('Unemployment Rate Over the Pandemic Years')
    
    # Display the graph
    plt.show()

        # Set the index as the x-axis for the first graph
    x1 = cleaned_data['index'].values
    y1 = cleaned_data.iloc[:, 1].values
    
    # Set the index as the x-axis for the second graph
    x2 = cleaned_data1['Unnamed: 0'].values
    y2 = cleaned_data1.iloc[:, 1].values
    
    # Plot the data as line graphs
    plt.plot(x1, y1, marker='o', label='Graph 1')
    plt.plot(x2, y2, marker='o', label='Graph 2')
    
    # Set the y-axis limits
    plt.ylim(0, max(max(y1), max(y2)) + 1)  # Add a buffer of 1 to the maximum value
    
    # Set the x-axis tick locations and labels
    plt.xticks(rotation=45)
    
    # Add labels and title
    plt.xlabel('Year')
    plt.ylabel('Unemployment Rate (%)')
    plt.title('Unemployment Rate Over the Years')
    
    # Add a legend
    plt.legend()
    
    # Display the graph
    plt.show()



# Run the main function
if __name__ == '__main__':
    main()


#Resposta: Como podemos analisar pelos grficos gerados, mesmo antes da pandemia a tendencia que se verificava era um decrescimo da taxa de desemprego.
#Podemos verificar que os anos pandmicos no afetaram essa tendncia em demasia pelo que a % continua a descer ainda a que um ritmo menos acelerado.
```
