```python
#Como a pandemia afetou os diversos eventos de cariz cultural?
#Para responder a esta questão analisamos dados relativos á lotação de diversos tipos de eventos culturais.
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker

# Function to load data
def load_data(filename):
    df = pd.read_excel(filename, skiprows=7)  # jumps the first 7 rows
    return df

# Function to clean data
def clean_data(df, label):
    df = df.iloc[:, :-2]  # Remove the last two columns
    row_index = df[df.columns[0]] == label
    data = df.loc[row_index]
    data = data.drop(data.columns[0], axis=1)
    data = data.transpose().reset_index()

    return data

# Main function
def main():
    filename = 'dados.xls'
    data = load_data(filename)
    cleaned_data_cinema = clean_data(data, 'Espetadores de cinema ')
    cleaned_data_museus = clean_data(data, 'Visitantes de museus ')
    cleaned_data_espetaculos = clean_data(data, 'Espetadores de espetáculos ao vivo ')

    # Convert pandas series to numpy array before plotting
    cinema = cleaned_data_cinema.iloc[:, 1].to_numpy()
    museus = cleaned_data_museus.iloc[:, 1].to_numpy()
    espetaculos = cleaned_data_espetaculos.iloc[:, 1].to_numpy()

    # Convert 'index' series to numpy array
    cinema_index = cleaned_data_cinema['index'].to_numpy()
    museus_index = cleaned_data_museus['index'].to_numpy()
    espetaculos_index = cleaned_data_espetaculos['index'].to_numpy()

    plt.plot(cinema_index, cinema, label='Cinema spectators')
    plt.plot(museus_index, museus, label='Museum visitors')
    plt.plot(espetaculos_index, espetaculos, label='Live show spectators')

    plt.xlabel('Year')
    plt.ylabel('Number of Spectators/Visitors')
    plt.title('Cultural Events Over the Years')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
     # Set the x-axis tick locations and labels
    plt.gca().xaxis.set_major_locator(ticker.MaxNLocator(integer=True))
    plt.show()

# Run the main function
if __name__ == '__main__':
    main()
#Resposta: Como podemos ver existiu uma queda abruta do numero de espetadores de eventos culturais durante os anos pandémicos.
```
